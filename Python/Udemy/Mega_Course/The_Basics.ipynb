<h3>Functions and Conditionals</h3>

```python
doit = [10,10,10]
def mean(values=None):
    the_mean = sum(values) / len(values)
    return the_mean
print(mean(doit))
mymean = mean(doit)
print(mymean + 10) 


def calculate_length(lst):
    return len(lst)
print(calculate_length(doit))

def square(a):
    return a * a
print(square(10))

def volume(oz):
    return oz * 29.57353
print(volume(10))

def dictmean(value=None):
    if type(value) == dict:
        the_mean = sum(value.values()) / len(value)
    else:
        the_mean = sum(value) / len(value)
    return the_mean
dictit = {"Jim" : 10, "Mary":9, "John":8}
print(dictmean(doit))
print(dictmean(dictit))

x = 1
y = 1
if x == 1 or y==2:
    print("Yes")
else:
    print("No")

x = 1
y = 1
if x == 1 and y==1:
    print("Yes")
else:
    print("No")

def foo(x, array):
    if x in array:
        return True
    else:
        return False

print(foo(1, [1, 2, 3]))
print(foo(1, [2, 3]))
print(foo(1, ['1', 2, 3]))

x = '3'
if isinstance(x, int) or isinstance(x, float) or x=='1':
    print("Valid type!")
else:
    print("Not valid!")

def passwordcheck(password):
    if len(password) >= 8:
        return True
    else:
        return False
print(passwordcheck('ajdjasjdajsdjasdj'))

def temp(temperature):
    if temperature > 65:
        return "Warm"
    else:
        return "Cold"
print(temp(80))

def high(temperature):
    if temperature > 25:
        return "Hot"
    elif temperature <= 25 and temperature >= 15:
        return "Warm"
    elif temperature < 15:
        return "Cold"
print(high(26))

```

<h3>Loops</h3>

```python
colors = [11, 34, 98, 43, 45, 54, 54]

for color in colors:
    if color > 50:
        print(color)

colors = [11, 34, 98, 43, 45, 54, 54]

for color in colors:
    print(color)

colors = [11, 34.1, 98.2, 43, 45.1, 54, 54]

for color in colors:
    if isinstance(color, int) and color > 50:
        print(color)

colors = [11, 34.1, 98.2, 43, 45.1, 54, 54]

for color in colors:
    if isinstance(color, int):
        print(color)

def celsius_to_kelvin(cels):
    return cels + 273.15
monday_temperatures = [9.1, 8.8, -270.15]
for temperature in monday_temperatures:
    print(celsius_to_kelvin(temperature))

phone_numbers = {"John Smith": "+37682929928", "Marry Simpons": "+423998200919"}
for pair in phone_numbers.items():
    print("{} has as phone number {}".format(pair[0], pair[1]))

phone_numbers = {"John Smith": "+37682929928", "Marry Simpons": "+423998200919"}
for key, value in phone_numbers.items():
    print(f"{key} has as phone number {value}")

phone_numbers = {"John Smith": "+37682929928", "Marry Simpons": "+423998200919"}

for value in phone_numbers.values():
    print(value.replace("+", "00"))

a = 3 
while a > 0:
    print(a + a)
username = ''
while username != "pypy":
    username = input("Enter Username:")

a = 0
while a < 5:
    a = a + 1
    print(a)

while True:
    username = input("Enter Username:")
    if username == 'pypy':
        break
    else:
        continue

import datetime
while datetime.datetime.now() < datetime.datetime(2090, 8, 20, 19, 30, 20):
    print("It's not yet 19:30:20 of 2090.8.20")
```
